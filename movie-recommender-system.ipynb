```python
# this above code will give the TypeError coz the data in the dataset is a string of 
# so we have to convert the string(list) to List
# to do that there is a in-built library named ast 
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#  know we have to remove the spaces between the words coz while we are combinig 
# all of them into a single TAG if there is a space b/w the words of a same thing 
# then the words will become separate tags 
# EX: Sam Worthington there is a space in b/w they they will become 
# tag1 -> Sam
# tag2 -> worthington, so thats y we use this
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# to know y we done this see line 83
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# know we are converting the text into the vectors is know as TextVectorization
# we will combine all the words is known as BAG OF WORDS and check with the evry 
# single word in the BAG OF WORDS and count how many times that the word is 
# repeated in the vector then that vector will be the closest vector
# to do this we use the 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2000,stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
```

```python
# this above line will turns your tags text column into a numeric matrix of 
# word counts using Bag‑of‑Words.
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
# in this above output we can see that there are many words which refreed the same thing
# EX: loved,love,loving -> we will convert this all into a single word like
# love, love, love with the process called steaming
# this steaming will be doing by using the library named NLTK
```

```python
# to find the similarity's between the movies we are going to find the COSINE SIMILARITY
# coz this cosine similarity will find out the angle b/w the two vectors 
# if the angle is small then the similarity is higher 
# if the angle is longer then the movie is not similar to the user's input movie
# to do this follow below code
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
